```python
import numpy as np
import matplotlib.pyplot as plt

mu_lambdaCDM = np.genfromtxt('x_real_580_lambdaCDM.csv', delimiter=',')
mu_omegaCDM = np.genfromtxt('x_real_580_omegaCDM.csv', delimiter=',')
mu_CPL = np.genfromtxt('x_real_580_CPL.csv', delimiter=',')

mu_lambdaCDM_with_err = np.genfromtxt('x_real_580_lambdaCDM_with_err.csv', delimiter=',')
mu_omegaCDM_with_err = np.genfromtxt('x_real_580_omegaCDM_with_err.csv', delimiter=',')
mu_CPL_with_err = np.genfromtxt('x_real_580_CPL_with_err.csv', delimiter=',')

z = np.genfromtxt('z_obs.csv')
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,4))

for (ax, mu, mu_err, title) in zip(axs, (mu_lambdaCDM, mu_omegaCDM, mu_CPL), 
                (mu_lambdaCDM_with_err, mu_omegaCDM_with_err, mu_CPL_with_err),
                ('$\Lambda$CDM', '$\omega$CDM', 'CPL')):
    lower, middle, upper = np.quantile(mu, [0.05, 0.5, 0.95], axis=0)
    ax.plot(z, middle, color='k', label=title)
    ax.fill_between(z, lower, upper, color='k', alpha=0.1, label=title+' err')
    ax.plot(z, mu[0], color='b', label=title+' sample')
    ax.scatter(z, mu_err[0], s= 1, color='r', label=title + ' sample with $\Sigma_{obs}$')
    ax.set_title(title)

    ax.set_xlabel('redshift $z$')
    ax.set_ylabel('distance modulus $\mu$')
    ax.legend()

#plt.suptitle('Examples training samples from three dark energy models')
plt.legend()
plt.savefig('training examples from LambdaCDM, omegaCDM, CPL.png')

plt.show()
```
